```
def clean_text(text):
  import re
  import string
  import nltk

  from nltk.corpus import stopwords

  stopwords = stopwords.words('english')
  ps = nltk.PorterStemmer()

  text = "".join([word.lower() for word in text if word not in string.punctuation])

  # DONE: remove <p> using regex
  text = re.sub('[<p>]', '', text)
  # DONE: re.sub('<.{1,2}>', '', text)
  text = re.sub('<.{1,2}>', '', text)

  tokens = re.split('\W+', text)
  text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])

  return text
```

- TF-IDF value increases in proportion to the number of times a word appears in the document but is often offset by the frequency of the word in the corpus, which helps to adjust with respect to the fact that some words appear more frequently in general

- TF-IDF use two statistical methods: First is Term Frequency and the other is Inverse Document Frequency.

- Term frequency refers to the total number of times a given term t appears in the document doc against (per) the total number of all words in the document

- The inverse document frequency measure of how much information the word provides. It measures the weight of a given word in the entire document. IDF show how common or rare a given word is across all documents.

```
def train_model(features, labels):
  from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
  from sklearn.model_selection import train_test_split, RandomizedSearchCV
  from sklearn.neural_network import MLPClassifier
  from sklearn.metrics import classification_report, confusion_matrix
  from sklearn.inspection import permutation_importance
  from joblib import dump, load
  import pandas as pd
  import seaborn as sns
  import matplotlib.pyplot as plt

  # DONE: update or remove by trying to stratify
  X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
  #X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42, stratify=labels)

  model = MLPClassifier(activation='tanh',
                         alpha=0.05,
                         hidden_layer_sizes=(10,),
                         learning_rate='adaptive',
                         max_iter=1000,
                         solver='lbfgs')

  model.fit(X_train, y_train)

  dump(model, '/tmp/trainmodel.joblib')
  trained_clf = load('/tmp/trainmodel.joblib')

  # DONE: adding permutation importance with Tfidf
  # https://scikit-learn.org/stable/modules/permutation_importance.html
  # https://stackoverflow.com/questions/52972368/select-top-n-tfidf-features-for-a-given-document
  r = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2)

  for i in r.importances_mean.argsort()[::-1]:
    if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
      print(f"{tfidf_features[i]:<8}"
            f"{r.importances_mean[i]:.3f}"
            f" +/- {r.importances_std[i]:.3f}")
  print("\n")

  # GOOD / DONE: include permutation_importance plots
  # https://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html#feature-importance-based-on-feature-permutation
  importances = pd.Series(r.importances_mean, index=tfidf_features).sort_values(ascending=False)
  fig, ax = plt.subplots(figsize=(20, 10))
  importances[:50].plot.bar(yerr=r.importances_std[:50], ax=ax)
  ax.set_title("Top Important TFIDF Features using Permutation Importance")
  ax.set_ylabel("Mean accuracy decrease")
  #fig.tight_layout()
  plt.show()

  # https://scikit-learn.org/stable/auto_examples/inspection/plot_permutation_importance.html#sphx-glr-auto-examples-inspection-plot-permutation-importance-py
  #sorted_importances_idx = r.importances_mean.argsort()
  #importances = pd.DataFrame(r.importances[sorted_importances_idx].T, columns=tfidf_features)
  #ax = importances.plot.box(vert=False, whis=10, figsize=(20, 10))
  #ax.set_title("Permutation Importances")
  #ax.axvline(x=0, color="k", linestyle="--")
  #ax.set_xlabel("Decrease in accuracy score / Model error increase when the feature was permutated")
  ##ax.figure.tight_layout()

  # evaluation metrics
  y_pred = model.predict(X_test)
  print("Precision:", precision_score(y_test, y_pred, average='weighted'))
  print("Recall:", recall_score(y_test, y_pred, average='weighted'))
  print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
  print("\n")

  target_names = ['Activities','Apps','Events','Physical Products','Services','Stores','Stories']

  # classification report
  print('CLASSIFICATION METRICS')
  print(classification_report(y_test, y_pred, labels=target_names))
  print("\n")

  # confusion matrix
  cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
  fig, ax = plt.subplots(figsize=(10,10))
  disp.plot(ax=ax)

  model.fit(features, labels)

  dump(model, '/tmp/finalmodel.joblib')
  final_clf = load('/tmp/finalmodel.joblib')

  return final_clf, trained_clf


final_model, trained_model = train_model(features, labels)
```

### permutation_importance for Feature Importance

- Permutation Importance shuffles a feature and sees how the model changes its prediction.
- It is a technique for calculating relative importance scores that is independent of the model used.
- First, a model is fit on the dataset, such as a model that does not support native feature importance scores.
- Then the model is used to make predictions on a dataset, although the values of a feature (column) in the dataset are scrambled.
- This is repeated for each feature in the dataset. Then this whole process is repeated 3, 5, 10 or more times.
- The result is a mean importance score for each input feature (and distribution of scores given the repeats)
- Permutation Importance measures the importance of a feature by calculating the increase in the model’s prediction error after permuting the feature.
- A feature is “important” if shuffling its values increases the model error, because in this case the model relied on the feature for the prediction.
- A feature is “unimportant” if shuffling its values leaves the model error unchanged, because in this case the model ignored the feature for the prediction.
